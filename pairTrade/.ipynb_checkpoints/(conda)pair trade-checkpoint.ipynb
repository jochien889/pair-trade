```python
def get_bitfinex_asset(asset, ts_ms_start, ts_ms_end):
    url = 'https://api.bitfinex.com/v2/candles/trade:1D:t' + asset + '/hist'
    params = { 'start': ts_ms_start, 'end': ts_ms_end, 'sort': 1}
    r = requests.get(url, params = params)
    data = r.json()
    return pd.DataFrame(data)[2]
```

```python
start_date = 1514768400000 # 1 January 2018, 00:00:00
end_date = 1527811199000   # 31 May 2018, 23:59:59
assets = ['BTCUSD', 'ETHUSD', 'LTCUSD', 'XMRUSD', 'NEOUSD', 'XRPUSD', 'ZECUSD']
```

```python
def z_score(A, B):
    A_return = A.pct_change()
    B_return = B.pct_change()
    spread = A_return - B_return

    print(spread.mean())
#     print(spread.std())
    return (spread.iloc[-1] - spread.mean()) / spread.std(), spread.mean(), 

def upToOut(aPrice, bPrice):
    """
    statusList = (1,0)
    """
    balance['A']['available'] += (aPrice - aPastOrderPrice) * balance['A']['position'] + 1000
    balance['A']['position'] -= balance['A']['position']
    balance['B']['available'] += (bPrice - bPastOrderPrice) * balance['B']['position'] +1000
    balance['B']['position'] -= balance['B']['position']  
    
def upToDown(aPrice, bPrice):
    """
    statusList = (1,-1)
    """
    balance['A']['available'] += (aPrice - aPastOrderPrice) * balance['A']['position'] + 1000
    balance['A']['position'] -= balance['A']['position']
    balance['B']['available'] += (bPrice - bPastOrderPrice) * balance['B']['position'] +1000
    balance['B']['position'] -= balance['B']['position']    

    balance['A']['available'] -=  aPastOrderPrice 
    balance['A']['position'] += mergin/aPrice
    balance['B']['available'] -= bPastOrderPrice
    balance['B']['position'] -= mergin/bPrice 
    
def outToUp(aPrice, bPrice):
    """
    statusList = (0,1)
    """
    balance['A']['available'] -= mergin 
    balance['A']['position'] += -(mergin/aPrice)
    balance['B']['available'] -= mergin
    balance['B']['position'] += mergin/bPrice
    
def outToDown(aPrice, bPrice):
    """
    statusList = (0,-1)
    """
    balance['A']['available'] -=  mergin 
    balance['A']['position'] += mergin/aPrice
    balance['B']['available'] -= mergin
    balance['B']['position'] += -(mergin/bPrice)  
    
def downToOut(aPrice, bPrice):
    """
    statusList = (-1,0)
    """
    balance['A']['available'] += (aPrice - aPastOrderPrice) * balance['A']['position']
    balance['A']['position'] -= balance['A']['position']
    balance['B']['available'] += (bPrice-bPastOrderPrice) * balance['B']['position']
    balance['B']['position'] -= balance['B']['position']  
    
def downToUp(aPrice, bPrice):
    """
    statusList = (-1,1)
    """
    balance['A']['available'] += (aPrice - aPastOrderPrice) * balance['A']['position']
    balance['A']['position'] -= balance['A']['position']
    balance['B']['available'] += (bPrice-bPastOrderPrice) * balance['B']['position']
    balance['B']['position'] -= balance['B']['position']  
    
    balance['A']['available'] -=  aPastOrderPrice 
    balance['A']['position'] -= mergin/aPrice
    balance['B']['available'] -= bPastOrderPrice
    balance['B']['position'] += mergin/bPrice


strategy = {(1,0): upToOut, 
            (1,-1):upToDown,
            (0,1): outToUp,
            (0,-1):outToDown,
            (-1,0):downToOut,
            (-1,1):downToUp}    
```

```python
# Trading Ratio

import numpy as np
import pandas as pd
import os
# os.chdir('D:/Pair_Trading/參考資料/PythonCodesAndData/PythonCodesAndData')
# df=pd.read_csv('inputData_GLD_USO.csv')
symbols = ['BTCUSDT', 'ETHUSDT']
df = ((pd.concat([pd.read_csv(f"crypto_ohlc/{symbol}.csv", index_col='date', parse_dates=True)
        ['close'].rename(symbol)for symbol in symbols],axis=1,sort=True)))

# df['Date']=pd.to_datetime(df['Date'],  format='%Y%m%d').dt.date # remove HH:MM:SS
# df.set_index('Date', inplace=True)

lookback=20
ratio= df['BTCUSDT']/df['BTCUSDT'].shift(1) - df['ETHUSDT']/df['ETHUSDT'].shift(1)
#ratio= df['USO']/df['GLD']
ratio.plot()


#yport=np.sum(ts.add_constant(-hedgeRatio)[:, [1,0]]*np.log(df), axis=1)
#yport.plot()

# Apply a simple linear mean reversion strategy to GLD-USO
numUnits =-(ratio-ratio.rolling(lookback).mean())/ratio.rolling(lookback).std() # capital invested in portfolio in dollars.  movingAvg and movingStd are functions from epchan.com/book2
positions=pd.DataFrame(np.tile(numUnits.values, [2, 1]).T * np.ones((numUnits.shape[0], 2)) * np.array([-1, 1]) ) # positions in dollar invested
pnl=np.sum((positions.shift().values)*(df.pct_change().values), axis=1) # daily P&L of the strategy
ret=pnl/np.sum(np.abs(positions.shift()), axis=1)
# (np.cumprod(1+ret)-1).plot()
kk = pd.DataFrame((np.cumprod(1+ret)))
kk.index = ratio.index
kk.plot()
print('APR=%f Sharpe=%f' % (np.prod(1+ret)**(252/len(ret))-1, np.sqrt(360)*np.mean(ret)/np.std(ret)))
```

```python
# Bollinger Band Mean Reversion Strategy
import os
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import statsmodels.tsa.stattools as ts
#import statsmodels.tsa.vector_ar.vecm as vm


# import yfinance as yf  # yfinance 取得股價

# os.chdir('D:\\Pair_Trading\\crypto_ohlc')
df1 = pd.read_csv('crypto_ohlc/BTCUSDT.csv')['close']
df2 = pd.read_csv('crypto_ohlc/ETHUSDT.csv')['close']

df1 = df1.head(30)
df2 = df2.head(30)
#df1 = yf.download("uso", "2006-04-26")['Close']
#df2 = yf.download("gld", "2006-04-26")['Close']
df1.name = "BTC"
df2.name = "ETH"
df3 = pd.concat([df2, df1], axis=1)
#df3.to_csv("GLD-USO-max.csv")

#df3.index = pd.Series(df3.index).apply(lambda x: x.to_pydatetime().date())

#type(df3.index)
#type(df.index[0])

df=df3

lookback=20
hedgeRatio=np.full(df.shape[0], np.nan)
e = []
for t in np.arange(lookback, len(hedgeRatio)):
    regress_results=sm.ols(formula="BTC ~ ETH", data=df[(t-lookback):t]).fit() # Note this can deal with NaN in top row
    hedgeRatio[t-1]=regress_results.params[1] ## beta1
    e.append(regress_results.params[0])
    
    
yport=np.sum(ts.add_constant(-hedgeRatio)[:, [1,0]]*df, axis=1)
yport.plot()

# Bollinger band strategy
entryZscore=1
exitZscore=0

MA=yport.rolling(lookback).mean()
MSTD=yport.rolling(lookback).std()
zScore=(yport-MA)/MSTD

longsEntry=zScore < -entryZscore # -1
longsExit =zScore > -entryZscore # -1

shortsEntry=zScore > entryZscore # 1
shortsExit =zScore < exitZscore  # 0

numUnitsLong=np.zeros(longsEntry.shape)
numUnitsLong[:]=np.nan

numUnitsShort=np.zeros(shortsEntry.shape)
numUnitsShort[:]=np.nan

numUnitsLong[0]=0
numUnitsLong[longsEntry]=1
numUnitsLong[longsExit]=0
numUnitsLong=pd.DataFrame(numUnitsLong)
numUnitsLong.fillna(method='ffill', inplace=True)

numUnitsShort[0]=0
numUnitsShort[shortsEntry]=-1
numUnitsShort[shortsExit]=0
numUnitsShort=pd.DataFrame(numUnitsShort)
numUnitsShort.fillna(method='ffill', inplace=True)

numUnits=numUnitsLong+numUnitsShort
positions=pd.DataFrame(np.tile(numUnits.values, [1, 2]) * ts.add_constant(-hedgeRatio)[:, [1,0]] *df.values) #  [hedgeRatio -ones(size(hedgeRatio))] is the shares allocation, [hedgeRatio -ones(size(hedgeRatio))].*y2 is the dollar capital allocation, while positions is the dollar capital in each ETF.
pnl=np.sum((positions.shift().values)*(df.pct_change().values), axis=1) # daily P&L of the strategy
ret=pnl/np.sum(np.abs(positions.shift()), axis=1)
cumret = pd.DataFrame((np.cumprod(1+ret)))

# print('APR=%f Sharpe=%f' % (np.prod(1+ret)**(252/len(ret))-1, np.sqrt(252)*np.mean(ret)/np.std(ret)))
```
